Welcome to the survival game. This game will test your survival knowledge. 
The game will test you in a random scene with a selected random inventory.
There is only one correct answer for a question, if you answer correctly you will advance If not the game will automatically restart. 

Stage 1. First decision. Water, shelter and food.
Stage 2. Second decision. How to get water.
Stage 3. Third decision. Items for condensation water system.
Stage 4. Fourth decision. Food, shelter, SOS sign.
Stage 5. Fifth decision. Where to build your shelter.

Good Luck and enjoy!

```python
# import packages random, time and sys
import random as rd
import time 
import sys

# Asking the user his name and some other questions

user_name = input(prompt = "What is your name? (Name)\n\t > ")
user_country = input("Where are you from? (Country)\n\t > ")
experience = input("Do you have any survival experience? (Yes or No)\n\t > ")

# defining some variables important for the game

scenarios = 'an island'
inventory = ['1 liter of water','backpack']
backpack1 = ['multitool knife', 'pocket knife', 'survival knife']
backpack2 = ['matches', 'lighter', 'pocket knife','fire starter']
backpack3 = ['sleeping bag', 'first aid kit','compass']

# Defining function to print loading game using a loop to type letter by letter.
def loading_game():
    load = "LOADING GAME ..."
    """function use to spell letter by letter as a writting machine""" 
    for letter in load:
        print(letter, end ="", flush = True)
        time.sleep(0.030)

# Generating random inventory & introduction
def random_items():
    """function defininf random items in inventory and Introduction for the game"""
    inventory.append(rd.choice(backpack1))
    inventory.append(rd.choice(backpack2))
    inventory.append(rd.choice(backpack3))
    
    loading_game()
    
# making the text appear in a few seconds
    time.sleep(3)

# printin dynamic introduction of the game
    print(
    f"""
    \n\t{user_name} was returning to {user_country} by plane coming from
    \tAustralia. The flight was going well. It was unexpected that   
    \thalfway through the flight, the captain asks the passengers to 
    \tfasten their seat belts. {user_name} was sitting in a privileged 
    \t location. {user_name} was able to see the flight attendants.
    """)
    
    time.sleep(8)
    
    print(
    f"""
    \n\t{user_name} knew that something was wrong, the flight attendant 
    \tface said it all.
    """)
    
    time.sleep (5)
    
    print(
    f"""
    \n\tThe Captain announced the words you never want to hear when 
    \tflying.
    \n\t\t\t****** BRACE FOR IMPACT ******
    """)
    
    time.sleep(5)
    
    print(
    f"""
    \n\t{user_name} lost consciousness for a few hours to wake up on 
    \t{scenarios}, imediately finding a backpack with 
    \n\t{inventory}.
    \n\tIt is approximately noon, the next few hours will be critical  
    \tfor your survival.
    """)
    
    input(prompt = "<Press enter to continue>")
    
    time.sleep(3)
    
    loading_game()
    
random_items()

# defining fail function and then restarting the game if user wants to play again
def fail():
    """Asking the user if they want to play again"""
    fail_answer = input(prompt = """
    Do you want to play again?
    
    (Answer using yes or no)
    
    """)
    if fail_answer == "yes" or fail_answer == "y" or fail_answer == "ye":
        random_items()
        restart_game()
    else:
        sys.exit(0)
    
# defininf a function that restart the game and run it again
def restart_game():
    """Function that restart the game and run it again"""
    answer_r1 = input(prompt = 
    f"""
    \n\tYou realized you are stock on an island, what would you do as
    \tyour first survival decision:
    
    a. water
    b. shelter
    c. food
    
    """)
    if answer_r1 == "a" or answer_r1 == "a." or answer_r1 =="wather" or answer_r1 == "water":
        print(
        f"""
        \n\tOn track to survive day 1.
        """)
        day1b_island()
        day1c_island()
        day2a_island()
        day2b_island()
    elif answer_r1 == "b" or answer_r1 == "b." or answer_r1 == "shelter":
        print(
        f"""
        \n\tOne bottle of water is not enought to survive a hot day. 
        \tGame Over
        """)
        random_items()
        day1a_island()
        day1b_island()
        day1c_island()
        day2a_island()
        day2b_island()
    elif answer_r1 == "c" or answer_r1 == "c." or answer_r1 == "food":
        print(
        f"""
        \n\tWhat are you thinking? food before water?.
        \tGame Over
        """)
        random_items()
        day1a_island()
        day1b_island()
        day1c_island()
        day2a_island()
        day2b_island()
    else:
        print("Game Over")
        random_items()
        day1a_island()
        day1b_island()
        day1c_island()
        day2a_island()
        day2b_island()

# defining a function for the winners of the game       
def won_game():
    """function that shows that you won the game"""
    print(
    f"""
    --------------------------------------------------------------------
    \t\tCongratulations! You Won!!
    --------------------------------------------------------------------
    """)

    time.sleep(5)
    sys.exit()
    
# defining question 1a Island (stage1)
def day1a_island():
    """Ask questions and answers, check if proceed or restart game"""
    answer1 = input(prompt = 
    f"""
    \n\tYou realized you stock on an island, what would you do as
    \tyour first survival decision:
    
    a. water 
    b. shelter
    c. food
    
    """)
    if answer1 == "a" or answer1 == "a." or answer1 == "water" or answer1 == "wather":
        print(
        f"""
        \n\tOn track to survive day1."
        """)
        
    elif answer1 == "b" or answer1 == "b." or answer1 == "shelter" or answer1 == "sheltre":
        print(
        f"""
        \n\tWhat are you thinking? food before water?. 
        \tGame Over
        """)
        fail()
        
    elif answer1 == "c" or answer1 == "c." or answer1 == "food" or answer1 == "foood":
        print(
        f"""
        \n\tWhat are you thinking? food before water?. 
        \tGame Over
        """)
        fail()
    else:
        print("Game Over")
        fail()
        
        time.sleep(3)

    loading_game()

day1a_island()    
        
# defining question 1b Island (stage2)

def day1b_island():
    """Ask questions and answers, check if proceed or restart game"""
    answer2 = input(prompt = 
    f"""
    \n\tThere are 3 options for a water source, which one would you pick?
    
    a. natural 
    b. coconuts
    c. fabricate condensation device
    
    """)
    
    if answer2 == "a" or answer2 == "a." or answer2 == "natural":
        print(
        f"""
        \n\tWhile you were looking for a water source, you fall down
        \tinto a deep hole, where you died.
        \tGame Over
        """)
        fail()
        
    elif answer2 == "b" or answer2 == "b." or answer2 == "coconuts" or answer2 == "coco":
        print(
        f"""
        \n\tThe sun and effort put in getting the coconuts took 
        \tall of your energy and you died.
        \tGame Over
        """)
        fail()
    elif answer2 == "c" or answer2 == "c." or answer2 == "fabricate condensation device" or answer2 == "condensation" or answer2 == "device":
        print(
        f"""
        \n\tNice job, on track to survive the first day.
        """)
    else:
        print(
        f"""
        \tGame Over.
        """)
        fail()

        time.sleep(3)
        
    loading_game()

day1b_island()

# defining question 1c Island (stage 3)

def day1c_island():
    """Ask questions and answers, check if proceed or restart game"""
    answer3 = input(prompt = 
    f"""
    \n\tWhat materials do you need to be able to collect water by condensation?
    
    a. plastic bag, flask, rock
    b. rubber ban, tin can, wooden stick
    c. fabric piece, container, rock
    
    """)

    if answer3 == "b" or answer3 == "b." or answer3 == "rubber ban" or answer3 == "tin can" or answer3 == "wooden stick":
        print(
        f"""
        \n\tGame Over
        """)
        fail()
    elif answer3 == "c" or answer3 == "c." or answer3 == "fabric piece" or answer3 == "container" or answer3 == "rock":
        print(f"""
        \n\tGame Over
        """)
        fail()
    elif answer3 == "a" or answer3 == "a." or answer3 == "a" or answer3 == "plastic bag" or answer3 == "flask" or answer3 == "rock":
        print("Nice job, you survive the first day.")
    else:
        print("Game Failure.")
        fail()

        time.sleep(3)
        
    loading_game()

day1c_island()

# defining intro for day2 
def day2_intro():
    """Function that introduces day n2"""
    time.sleep(3)
    
    loading_game()
    
    print(
    f"""
    \n\t{user_name} you survived day 1. The night was awful since there was no
    \ttime to set a fire, make a shelter, or gather some food. You did not sleep 
    \tmuch and the energy level is super low. Your life depends on today's decisions. 
    \tGood Luck
    """)
    
    time.sleep(3)
        
    input(prompt = "<Press any enter to continue>")
        
day2_intro()

# defining day2a first question (stage 4)

def day2a_island ():
    """Ask questions and answers, check if proceed or restart game with nested if that
    allows to win the game"""
    answer4 = input(prompt = 
    f"""
    \n\tWhat is your first decision in day 2?
    
    a. shelter
    b. food
    c. sos sign
    
     """)
    
    if answer4 == "a" or answer4 == "a." or answer4 == "shelter":
        print(
        f"""
        \n\tYour energy is to low, while building your shelter you broke your foot, leading to a 
        \tslow death. 
        \tGame Over
        """)
        fail()
    elif answer4 == "c" or answer4 == "c." or answer4 == "sos" or answer4 == "sign":
        bonus_question = input(prompt = 
        f"""
        \n\tWhat kind of sign would you do?
        
        a. fire 
        b. sos on the sand
        c. note inside the bottle
        """)
        
        if bonus_question == "a" or bonus_question == "a." or bonus_question == "fire":
            print(
            f""" 
            \n\tA boat nearby saw the fire sign and save you.
            \tyou Won!
            """)
            won_game()
        else:
            print(
            f"""
            \n\tWaste all your reamining energy, your body collapase. 
            \tGame Over
            """)
            fail()
    elif answer4 == "b" or answer4 == "b." or answer4 == "food" or answer4 == "foood":
        print(
        f"""
        \n\tGood Job! Food is esential to finish this day.
        """)
    else:
        print("Game Failure.")
        fail()

        time.sleep(3)
    
    loading_game()

day2a_island()

# defining day2 second question (Stage 5)

def day2b_island():
    
    answer5 = input(prompt = 
    f"""
    \n\tIf you still here, you missed the boat that passed few hours ago. 
    \tBut at least you gather some food. You turn on a fire and cook the snails
    \tyou collected from the ocean, also made a sign that is ready to be light on.
    
    \n\tNow you just started thinking about how to spend the night, where would 
    \tyou build your shelter?
    
    a. beach on the floor
    b. top of a tree
    c. inside a cave
    """)
    if answer5 == "a" or answer5 == "a." or answer5 == "bea" or answer5 == "flo" or answer5 =="th":
        print(
        f"""
        \n\tYour were eaten by a panther. 
        \tGame Over
        """)
        fail()
    elif answer5 == "b" or answer5 == "b." or answer5 == "to" or answer5 == "tre":
        b_question2 = input(prompt = 
        f"""
        \n\tHeard a boat while sleeping, woke up. Need to send some signal to the boat.
        \tWhat would you do?
        
        a. dancing
        b. lighting sos sign
        c. screaming
        """)
        
        if b_question2 == "b" or b_question2 == "b." or b_question2 == "lig" or b_question2 == "so":
            print(""" A boat nearby saw the fire sign and save you, you Won!""")
            won_game()
        else:
            print(
            f"""
            \n\tWaste all your reamining energy, your body collapase.
            \tGame Over""")
        fail()
    elif answer5 == "c" or answer5 == "c." or answer5 == "ins" or answer5 == "cav":
        print(
        f"""
        \n\tGood Job! you survive the night, however the tide did not allow you 
        \tto signal the nearby boat. However another boat rescue you next morning
        """)
        won_game()
    else:
        print(
        f"""
        \n\tGame Failure.
        """)
        fail()

        time.sleep(3)
    
    loading_game()
    
day2b_island()
```

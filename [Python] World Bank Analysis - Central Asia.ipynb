```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# importing file
file = "./datasets/Final_Project_Dataset.xlsx"

# reading excel in python
final_project_datasets = pd.read_excel(io         = file,
                                       data_sheet = "Data",
                                       header     = 0)

# display max columns and max rows
pd.set_option("display.max_columns", None)

# filtering for Central Asia
central_asia = final_project_datasets.loc[ : , : ] \
              [final_project_datasets.loc[:,"Hult Region"] == "Central Asia"]

central_asia
```

```python
# descriptive statistics for variables
central_asia.describe().round(decimals = 2)
```

```python
# filling NaN wit mean
central_asia_fill = central_asia.fillna(central_asia.mean())

# correlation between variables
central_asia_fill_corr = central_asia_fill.corr().round(2)

#printing correls
#print(central_asia_fill_corr)
```

```python
# correlationship matrix

# utilizing palplot to create a color map
sns.palplot(sns.color_palette('mako', 12))

#plot size
fig, ax = plt.subplots(figsize=(25,25))


#creating heat map
sns.heatmap(central_asia_fill_corr,
            cmap = 'mako',
            square = False,
            annot = True,
            linecolor = 'white',
            linewidths = 0.5,
            cbar = False)

#editing the map
bottom, top = plt.ylim()
bottom += 0.5            
top -= 0.5               
plt.ylim(bottom, top)    

#saving the figure and displaying the plot
plt.savefig('Diamond Correlation Heatmap All Variables.png')
plt.show()

```

```python
# defining our 5 variables, we picked the ones more correlated to GDP.
central_asia_five = central_asia_fill.iloc[:,[1,9,15,18,22,33,38]]
central_asia_five
```

```python
# renaming lables of columns
central_asia_five_r = central_asia_five.rename(columns={
    'CO2 emissions (metric tons per capita)'       : 'CO2 Emissions',
    'Employment to population ratio, 15+, total (%) (modeled ILO estimate) ': 'Employment to population %',
    'GDP per person employed (constant 2011 PPP $)' : 'GDP',
    'Improved sanitation facilities (% of population with access)': 'Improved Sanitation Facilities %',
    'Population, total' : 'Population',
    'Primary completion rate, total (% of relevant age group)' : "Primary Completion Rate"
})
central_asia_five_r
```

```python
# correlation between variables
central_asia_five_r_corr = central_asia_five_r.corr().round(2)

# correlationship matrix

# utilizing palplot to create a color map
sns.palplot(sns.color_palette('mako', 5))

#plot size
fig, ax = plt.subplots(figsize=(5,5))


#creating heat map
sns.heatmap(central_asia_five_r_corr,
            cmap = 'mako',
            square = False,
            annot = True,
            linecolor = 'white',
            linewidths = 0.5,
            cbar = False)

#editing the map
bottom, top = plt.ylim()
bottom += 0.5            
top -= 0.5               
plt.ylim(bottom, top)    

#saving the figure and displaying the plot
plt.savefig('Corellation Heatmap 1 Selected Variables.png')
plt.show()


```

```python
# group N2 (tourism industry)
central_asia_group2 = central_asia_five_r.iloc[[6, 7, 9, 11],:]
central_asia_group2
```

```python
# converting correlation matrix into a DataFrame
central_asia_group2_corr = central_asia_group2.corr(method = 'pearson').round(decimals = 2)

# instantiating a heatmap
sns.heatmap(data = central_asia_group2.corr(method = 'pearson').round(decimals = 2),
            cmap = "mako")                  # changing to inferno colors


# displaying the plot
plt.show()

# specifying plot size (making it bigger)
fig, ax = plt.subplots(figsize=(6,6))


# developing a spicy heatmap
sns.heatmap(data       = central_asia_group2_corr, # the correlation matrix
            cmap       = 'mako',                   # changing to inferno colors
            square     = True,                     # tightening the layout
            annot      = True,                     # should there be numbers in the heatmap
            linecolor  = 'white',                  # lines between boxes
            linewidths = 0.5)                      # how thick should the lines be?

# title and displaying the plot
plt.title("""
Linear Correlation Heatmap for Sri Lanka, Maldives, 
Mongolia & Thailand
""")

plt.show()
```
